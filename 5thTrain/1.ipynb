```python
#board = [list(input().strip()) for i in range(8)]
board =[list('********'),
        list('*R******'), 
        list('********'), 
        list('*****B**'), 
        list('********'), 
        list('********'), 
        list('********'), 
        list('********')]
k = 64
for i in range(8):
    for j in range(8):
        if board[i][j] == 'B':
            k -= 1
            for h in range(1, min(i,j)+1):
                if board[i-h][j-h] == '*':
                    board[i-h][j-h] = 1
                    k -= 1
                elif board[i-h][j-h] == 1: continue
                else: break
            for h in range(1, min(8-i,8-j)):
                if board[i+h][j+h] == '*':
                    board[i+h][j+h] = 1
                    k -= 1
                elif board[i+h][j+h] == 1: continue
                else: break
            for h in range(1, min(i+1,8-j)):
                if board[i-h][j+h] == '*':
                    board[i-h][j+h] = 1
                    k -= 1
                elif board[i-h][j+h] == 1: continue
                else: break
            for h in range(1, min(8-i,j+1)):
                if board[i+h][j-h] == '*':
                    board[i+h][j-h] = 1
                    k -= 1
                elif board[i+h][j-h] == 1: continue
                else: break
        if board[i][j] == 'R':
            k -= 1
            for h in range(1, i+1):
                if board[i-h][j] == '*':
                    board[i-h][j] = 1
                    k -= 1
                elif board[i-h][j] == 1: continue
                else: break
            for h in range(1, j+1):
                if board[i][j-h] == '*':
                    board[i][j-h] = 1
                    k -= 1
                elif board[i][j-h] == 1: continue
                else: break
            for h in range(1, 8-i):
                if board[i+h][j] == '*':
                    board[i+h][j] = 1
                    k -= 1
                elif board[i+h][j] == 1: continue
                else: break
            for h in range(1, 8-j):
                if board[i][j+h] == '*':
                    board[i][j+h] = 1
                    k -= 1
                elif board[i][j+h] == 1: continue
                else: break
print(k)
```

```python
n = int(input())
num = list(map(int, input().split()))
str = ''
s = num[0] % 2
for i in range(1, n):
    if (s * num[i]) % 2 != 0:
        s = 1
        str += 'x'
    elif (s + num[i]) % 2 != 0:
        s = 1
        str += '+'
    else:
        s = 0
        str += '+'
print(str)
```

```python
n = int(input())
nums = list(map(int, input().split()))
str = ''
state = 'first'
for num in nums:
    if state == 'first':
        if num % 2 == 0:
            str += '+'
        else:
            state = 'second'
    elif state == 'second':
        if num % 2 == 0:
            str += '+'
            state = 'third'
        else:
            str += 'x'
    elif state == 'third':
        str += 'x'
print(str)
```

```python
my_units = int(input())
house = int(input())
spawn = int(input())
ans = 10**9
for t in range(house+1):
    x = my_units
    hp = house
    count = 0
    y = 0
    while hp >= t:
        if y >= x: 
            count = 10**9
            y = 0
            break
        hp -= x - y
        y = 0
        if hp > 0: y += spawn
        count += 1
    while hp > 0:
        if x <= 0: 
            count = 10**9
            y = 0
            break
        if hp >= x: hp -= x
        else:
            y -= x - hp
            hp = 0
        x -= y
        if hp > 0: y += spawn
        count += 1
    while y > 0:
        if x <= 0: 
            count = 10**9
            break
        y -= x
        if y > 0: x -= y
        count += 1
    if count < ans: ans = count
if ans < 10**9: print(ans)
else: print(-1)
```

```python
l, x1, v1, x2, v2 = map(int, input().split())
ans = 10**30
for i in range(2):
    dx = (x2 - x1 + l) % l
    dv = v1 - v2
    if dv < 0:
        dv = -dv
        dx = (-dx) % l
    if dx == 0: ans = 0
    if dv != 0:
        ans = min(ans, dx/dv)
    x2 = (-x2) % l
    v2 = -v2
if ans < 10**30:
    print('yes')
    print(ans)
else: print('no')
```
